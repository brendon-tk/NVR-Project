# Data Preparation

```python
import pandas as pd
import numpy as np

# 0. Convert to datetime and sort so rolling stats work properly
df['Batch Date'] = pd.to_datetime(df['Batch Date'], errors='coerce')
df = df.sort_values(['Product Name', 'Batch Date'])

# 1. Feature Engineering
# Rolling stats on spending (per product, over time)
df['Rolling_Mean_Spending_7d'] = df.groupby('Product Name')['Invoice Line Amount'].transform(lambda x: x.rolling(window=7, min_periods=1).mean())
df['Rolling_Std_Spending_7d'] = df.groupby('Product Name')['Invoice Line Amount'].transform(lambda x: x.rolling(window=7, min_periods=1).std())
df['Rolling_Mean_Spending_30d'] = df.groupby('Product Name')['Invoice Line Amount'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
df['Rolling_Std_Spending_30d'] = df.groupby('Product Name')['Invoice Line Amount'].transform(lambda x: x.rolling(window=30, min_periods=1).std())

# Create a binary column for critical problem categories
critical_categories = ['Framing', 'Recall/ASD']
df['Critical_Problem'] = df['Problem Category'].apply(lambda x: 1 if x in critical_categories else 0)

# Create a binary column for top 5 highest-spending products
top_products = df.groupby('Product Name')['Invoice Line Amount'].sum().nlargest(5).index
df['Top_Product'] = df['Product Name'].apply(lambda x: 1 if x in top_products else 0)

# Count of tickets (encoded as 1s) for rolling ticket features
df['Ticket'] = 1
df['Rolling_Count_Tickets_7d'] = df.groupby('Product Name')['Ticket'].transform(lambda x: x.rolling(window=7, min_periods=1).sum())
df['Rolling_Count_Tickets_30d'] = df.groupby('Product Name')['Ticket'].transform(lambda x: x.rolling(window=30, min_periods=1).sum())

# 2. Data Splitting (time-based split, assuming df is sorted)
train_size = int(len(df) * 0.8)
df_train = df.iloc[:train_size].copy()
df_test = df.iloc[train_size:].copy()

# 3. Handle Missing Values (numeric columns only, impute with train mean)
num_cols = df_train.select_dtypes(include=[np.number]).columns
for col in num_cols:
    train_mean = df_train[col].mean()
    df_train[col] = df_train[col].fillna(train_mean)
    df_test[col] = df_test[col].fillna(train_mean)

# show results
display(df_train.head())
display(df_test.head())
```

# Data Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Select features
features_for_clustering = [
    'Rolling_Mean_Spending_7d', 'Rolling_Std_Spending_7d',
    'Rolling_Mean_Spending_30d', 'Rolling_Std_Spending_30d',
    'Rolling_Count_Tickets_7d', 'Rolling_Count_Tickets_30d',
    'Top_Product', 'Critical_Problem'
]
df_clustering = df_train[features_for_clustering].copy()

# 2. Fill missing values
for col in df_clustering.columns:
    if df_clustering[col].isnull().any():
        mean_value = df_clustering[col].mean()
        df_clustering.loc[:, col] = df_clustering[col].fillna(mean_value)

# 3. Scale features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clustering)

# 4. Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
df_train_clustered = df_train.copy()
df_train_clustered['cluster'] = kmeans.fit_predict(scaled_data)

# 5. Show results
display(df_train_clustered.head())
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import seaborn as sns

# --- STEP 1: prepare Prophet data ---
prophet_df = df_train_clustered[['Month', 'Invoice Line Amount']].rename(columns={'Month': 'ds', 'Invoice Line Amount': 'y'})

# --- STEP 2: run Prophet forecast ---
model = Prophet()
model.fit(prophet_df)

future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# --- STEP 3: resample to monthly for comparison ---
forecast_monthly = forecast.set_index('ds').resample('M').sum().reset_index()
forecast_monthly = forecast_monthly[['ds', 'yhat']]

monthly = prophet_df.set_index('ds').resample('M').sum().reset_index()
monthly.columns = ['ds', 'y']

# --- STEP 4: evaluate Prophet ---
y_true = monthly['y']
y_pred = forecast_monthly.set_index('ds').loc[monthly['ds']]['yhat'].values

print("MAE:", mean_absolute_error(y_true, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))

# --- STEP 5: calculate residuals and flag Prophet-based anomalies ---
monthly['yhat'] = y_pred
monthly['residual'] = monthly['y'] - monthly['yhat']
threshold = monthly['residual'].std() * 1.5
monthly['is_anomaly'] = monthly['residual'].abs() > threshold

# --- STEP 6: plot residuals ---
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly, x='ds', y='residual', label='Residual')
plt.axhline(threshold, color='red', linestyle='--', label='Threshold')
plt.axhline(-threshold, color='red', linestyle='--')
plt.title("Prophet Residuals with Anomaly Thresholds")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- STEP 7: create final anomaly flag in original data ---
# 7a. extract flagged months from Prophet
prophet_anomaly_months = set(monthly.loc[monthly['is_anomaly'], 'ds'].dt.to_period('M').astype(str))

# 7b. ensure month string column exists
df_train_clustered['month_str'] = df_train_clustered['Batch Date'].dt.to_period('M').astype(str)

# 7c. merge Isolation Forest + Prophet flags
df_train_clustered['final_anomaly'] = (
    (df_train_clustered['anomaly'] == -1) |
    (df_train_clustered['month_str'].isin(prophet_anomaly_months))
).astype(int)

# --- STEP 8: optional summary ---
print("Total Isolation Forest anomalies:", (df_train_clustered['anomaly'] == -1).sum())
print("Total Prophet-anomalous months:", len(prophet_anomaly_months))
print("Total final anomalies:", df_train_clustered['final_anomaly'].sum())

# check a few rows
display(df_train_clustered[df_train_clustered['final_anomaly'] == 1].head())
```
